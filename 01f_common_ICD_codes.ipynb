```python
import numpy as np
import pandas as pd
import pyreadstat
import simple_icd_10_cm as cm
```

```python
PCR_ICD_df = pd.read_csv("/Volumes/Research/GoldenHourData/InclusionCriteria/TESTfiltered_pcrkey_icdcodes_combined.csv")
PCR_ICD_df
```

```python
dx_columns = ['dx1', 'dx2', 'dx3', 'dx4', 'dx5']
all_dx_entries = pd.concat([PCR_ICD_df[dx] for dx in dx_columns])
entry_counts = all_dx_entries.value_counts()
top_10_entries = entry_counts.head(10)

print("Top 10 most common entries and their counts:")
print(top_10_entries)
```

```python
def icd_code_desc(icd_code):
    try: return cm.get_description(icd_code)
    except: return icd_code
```

```python
# convert icd codes to descriptions

icd_code_descriptions = pd.DataFrame({
    'ICD_Code': top_10_entries.index,
    'Count': top_10_entries.values,
    'Description': [icd_code_desc(code) for code in top_10_entries.index]
})
icd_code_descriptions
```
